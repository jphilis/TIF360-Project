```python
import openpyxl
import shutil
import os

# Define paths
excel_file = "/Users/adhvaitsrinath/Desktop/cleaning_dataset/eta.xlsx"
source_folder = "/Users/adhvaitsrinath/Desktop/cleaning_dataset/data_eta"
destination_folder = "/Users/adhvaitsrinath/Desktop/cleaning_dataset/organized_files"

# Load the Excel workbook
wb = openpyxl.load_workbook(excel_file)
sheet = wb.active  # Assuming data is in the first sheet

# Loop through each row (skip header row)
for row in sheet.iter_rows(min_row=2):
    # Get file path and scientific name
    file_path = row[16].value  # Assuming filename is in first column (16th index)
    scientific_name = row[9].value  # Assuming scientific name is in second column (9th index)
    
    # Check if file path is not None
    if file_path:
        # Extract base filename from the path (without extension)
        base_filename = os.path.splitext(os.path.basename(file_path))[0]

        # Search for matching file in source folder
        matching_file = None
        for filename in os.listdir(source_folder):
            if base_filename in filename:
                matching_file = filename
                break  # Stop iterating after first match

        # Handle cases with and without a match
        if matching_file:
            # Construct source and destination paths
            source_path = os.path.join(source_folder, matching_file)
            dest_path = os.path.join(destination_folder, scientific_name)

            # Create destination folder if it doesn't exist
            if not os.path.exists(dest_path):
                os.makedirs(dest_path)

            # Move the file
            shutil.move(source_path, dest_path)
            print(f"Moved '{matching_file}' to '{dest_path}'")
        else:
            print(f"No matching file found for '{base_filename}' in source. Skipping...")
    else:
        print("File path is None. Skipping...")
```
