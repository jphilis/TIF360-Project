## Code for renaming files to the scientific names

```python
import os
import pandas as pd
import logging

# Configure logging
logging.basicConfig(filename='rename_log.txt', level=logging.DEBUG, format='%(asctime)s %(message)s')

# Defining paths
folder_path = "/Users/adhvaitsrinath/Desktop/neural_network/data_folder"
excel_path = "/Users/adhvaitsrinath/Desktop/neural_network/Data.xlsx"
destination_folder = "/Users/adhvaitsrinath/Desktop/neural_network/renamed_files"

# Destination folder if it doesn't exist
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

# Excel file
try:
    df = pd.read_excel(excel_path)
except FileNotFoundError:
    logging.error(f"Error: Excel file not found at {excel_path}")
    exit()

# Create ID to name dictionary
id_to_name = dict(zip(df['tid'].astype(str).str.strip(), df['art3']))

# Dictionary mapping bat species names to their scientific names
# Received scientific names from excel
bat_species_to_scientific = {
    "Brunflagermus": "Myotis brandtii",
    "Sydflagermus": "Nyctalus leisleri",
    "Vandflagermus": "Myotis daubentonii",
    "Dværgflagermus": "Pipistrellus pipistrellus",
    "Brun-/Skimmel-/Sydflagermus": "Nyctalus noctula",
    "Troldflagermus": "Plecotus auritus",
    "Damflagermus": "Myotis dasycneme",
    "Frynseflagermus": "Myotis nattereri",
    "Myotis sp": "Myotis sp",
    "Skimmelflagermus": "Myotis bechsteinii",
    "Dam-/Vandflagermus": "Myotis dasycneme",
    "Bredøret Flagermus": "Plecotus auritus",
    "Brandts-/Bechsteins-/Skægflagermus": "Myotis brandtii / Myotis bechsteinii / Myotis mystacinus",
    "Støj": "Noise",
    "Ubestemt flagermus": "Undetermined bat",
    "Brun Langøre": "Plecotus austriacus"
}

# Iterating over the files to rename them
for filename in os.listdir(folder_path):
    if filename.endswith('.wav'):
        id_str = filename.split('_')[-2].strip()

        logging.info(f"Processing file: {filename}, ID from filename: {id_str}")

        if id_str in id_to_name:
            bat_species_name = id_to_name[id_str]
            scientific_name = bat_species_to_scientific.get(bat_species_name, bat_species_name)
            new_filename = filename.replace('.wav', f'_{scientific_name}.wav')

            try:
                os.rename(os.path.join(folder_path, filename), os.path.join(destination_folder, new_filename))
                logging.info(f'Renamed {filename} to {new_filename}')
            except FileNotFoundError:
                logging.error(f"Error: File not found {os.path.join(folder_path, filename)}")
        else:
            logging.warning(f'ID {id_str} not found in Excel file for {filename}')

```
